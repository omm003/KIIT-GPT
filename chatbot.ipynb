```python
from langchain_core.documents import Document
# CSV return in form of rows(tuples) and langchain split_documents accepts only in the form of documents
# Convert tuples (like ('id', None)) into Document objects
combined_dataset = [
    doc if isinstance(doc, Document)
    else Document(page_content=str(doc[1]), metadata={"source": str(doc[0])})
    for doc in combined_dataset
]
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 50
)

split_documents = text_splitter.split_documents(combined_dataset)

print(len(split_documents))
print(split_documents)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.llms import HuggingFacePipeline
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableParallel, RunnablePassthrough, RunnableLambda
from transformers import pipeline

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_store = FAISS.from_documents(split_documents, embeddings)
retriever = vector_store.as_retriever(search_kwargs={"k": 2})

llm_pipeline = pipeline(
    "text2text-generation",
    model="google/flan-t5-small",
    device=-1,
    max_new_tokens=256,
    truncation=True,
    temperature=0.6,
)
llm = HuggingFacePipeline(pipeline=llm_pipeline)

prompt = ChatPromptTemplate.from_template("""
Answer the user's question based only on the context below.
If the context does not have enough information, say "I don't have enough information."

Context:
{context}

Question:
{question}

Answer:
""")

def summarize_context(inputs):
    text = inputs["context"]
    if isinstance(text, list):
        text = " ".join([d.page_content for d in text])
    return {"context": text[:2000], "question": inputs["question"]}

rag_chain = (
    RunnableParallel({"context": retriever, "question": RunnablePassthrough()})
    | RunnableLambda(summarize_context)
    | prompt
    | llm
    | StrOutputParser()
)

query = "How are the placements at KIIT?"
result = rag_chain.invoke(query)
print("💬 Answer:", result)
```
